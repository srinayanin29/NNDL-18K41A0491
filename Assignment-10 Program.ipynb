```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import minmax_scale
from sklearn.model_selection import train_test_split
import seaborn as sns
from scipy import stats
```

**Read data**

```python
data=pd.read_excel("Load_Data_in_kW.xlsx",header=None)
data.head()
```

```python
data.info()
```

```python
z=np.abs(stats.zscore(data))
print(z)
```

```python
sns.boxplot(x=data[0])
```

```python
df=data
q=df[0].quantile(0.99)
df[df[0]<q]
q_low=df[0].quantile(0.01)
q_hi=df[0].quantile(0.99)
df_filtered=df[(df[0]<q_hi)&(df[0]>q_low)]
df_filtered.head()
```

```python
print(len(data))
print(len(df_filtered))
if(len(data)==len(df_filtered)):
  print('No outliers')
```

**Normalisation**

```python
data=df_filtered
data_n=data.copy()
data_n=(data_n-data_n.min())/(data_n.max()-data_n.min())
data_n.head()
```

**Split** **data**

```python
train_data,test_data=train_test_split(data_n,test_size=0.1)
print(len(train_data))
train_data.head()
```

```python
x=[]
y=[]
for i in range(len(train_data.iloc[:,0])-24):
  x.append(train_data.iloc[i,0])
  y.append(train_data.iloc[i+24,0])
```

**Train** **Data**

```python
m=1
c=-1
epochs=25  # epochs
l=0.1      # learning rate
ns=len(x)
gama=0.9
vm=0
vc=0
err=[]
for i in range(epochs):
  er=0
  for j in range(len(x)):
    er+=((y[j]-(m*x[j])-c))**2
    gm=((-1.0)*(y[j]-m*x[j]-c)*(x[j]))
    gc=((-1.0)*(y[j]-m*x[j]-c))
    vm=(gama*vm)-(l*gm)
    vc=(gama*vc)-(l*gc)
    m=m+vm
    c=c+vc
    er=(1/(2*len(x)))*(er)
    err.append(er)
print("The local minimum occurs at m =%.2f"%(m),", c = %.2f"%(c))
```

**RMSE for Training data**

```python
yp=[]
for i in range(len(x)):
  p=m*x[i]+c
  yp.append(p)
sum=0
for i in range(len(x)):
  sum+=(y[i]-yp[i])**2
  mse=sum/len(x)
print("Mean Square Error (MSE) : ",mse)
```

```python
plt.plot(np.array(x),np.array(yp),'green')
plt.scatter(np.array(x),np.array(y),color='red')
plt.title("Predicted data vs Actual data")
```

```python
iters=np.arange(epochs*len(x))
plt.plot(iters,err,'green')
plt.title("Error graph for training data")
plt.xlabel("iteration count")
plt.ylabel("error")
plt.grid()
```
